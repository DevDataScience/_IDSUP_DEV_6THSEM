```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3])
y = np.array([1, 3, 2])

plt.scatter(x, y)
```

```python
n = 1000
m= 0 
c= 0
lr= 0.1


for j in range(n):
  # y = mx + c 
  predicted_y = m*x + c  #numpy used for the accessing of list x

  #using sum of squares method
  slope_gradient = 0
  intercept_gradient = 0

  for i in range(3):
    residual = (y[i] - m * x[i] - c)  
    slope_gradient += -(2/3) * x[i] * residual
    intercept_gradient += -2/3 * residual

  m = m - lr * slope_gradient
  c = c - lr * intercept_gradient

  plt.plot(x, predicted_y)
```
